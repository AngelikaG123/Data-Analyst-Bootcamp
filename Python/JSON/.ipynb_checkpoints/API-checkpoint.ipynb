```python
import requests
import psycopg2
import json
import re

# insert database info
connection = psycopg2.connect(
    host=' ',
    user=' ',  
    password=' ', 
    dbname=' ', 
)

cursor = connection.cursor()

APIKEY1=' ' # insert API key to work with https://api.openweathermap.org
APIKEY2=' ' # insert API key to work with http://api.exchangeratesapi.io
```

```python
# From the database, select 10 customers with the highest creditLimit 
# and show it in currencies: USD, EUR, JPY and GBP.

cursor.execute('SELECT customername, creditLimit FROM customers ORDER BY creditLimit DESC LIMIT 10')

response = requests.get('https://api.exchangerate-api.com/v4/latest/USD')
data = response.json()

if response.status_code==200:
    for row in cursor:
        USD = round(row[1],2)
        EUR = round(row[1] * data['rates']['EUR'],2)
        JPY = round(row[1] * data['rates']['JPY'],2)
        GBP = round(row[1] * data['rates']['GBP'],2)
        result = f"{row[0]:30}: {USD:9} USD / {EUR:8} EUR / {JPY:10} JPY / {GBP:8} GBP"
        print(result)
else:
    print('Exchange rate table not found.')
```

```python
# Ask the user what currency he would like to convert PLN to.
# Then ask the user for the amount they would like to convert.

response = requests.get('http://api.exchangeratesapi.io/latest', {'access_key':APIKEY2})

result = 'Exchange PLN to:\n'
for i in response.json()['rates']:
    result = result + i + " "   
print(result)

currency = input("\nEnter your currency: ")
amount = int(input("\nEnter the amount of PLN to exchange: "))

if response.status_code==200:
    if currency in response.json()['rates']:
        EUR_to_PLN = response.json()['rates']['PLN']
        EUR = amount / EUR_to_PLN
        cash = round(EUR * response.json()['rates'][currency],2)
        print(f'{amount} PLN = {cash} {currency}')
    else:
        print("There is no such currency!")
else:
    print('No exchange rate found.')
    pass
```

```python
# Get the weather in London

response = requests.get(f'https://api.openweathermap.org/data/2.5/weather?q=London&appid={APIKEY1}')

if response.status_code == 401:
    print('Invalid key')
elif response.status_code == 200:
    print('Weather in London:', response.json()['weather'][0]['main'])
else:
    print('Different than expected error message!')
```

```python
# Get EUR-AUD exchange rate

response = requests.get(f'http://api.exchangeratesapi.io/latest?access_key={APIKEY2}')

if response.json()['success']:
    print('Exchange rate of EUR-AUD is now', response.json()['rates']['AUD'])
else:
    print('Error:', response.json()['error']['info'])
```

```python
# Get the weather in Warsaw with the help of a method with optional parameters 

def get_weather(city, state=None, country=None):    
    response = requests.get('https://api.openweathermap.org/data/2.5/weather', 
                            {'q':[city, state, country],'appid':APIKEY1})
    return response.json()['weather'][0]['main']

print('Weather in Warsaw:', get_weather('Warszawa', 'Mazowieckie', 'Polska'))
print('Weather in Warsaw:', get_weather('Warszawa'))
```

```python
# Check the weather in customers' locations. 
# If somewhere rains, give also company name, full name of a person of contact and phone number

cursor.execute('SELECT city FROM customers')

def get_weather(city, state=None, country=None):
    response = requests.get('https://api.openweathermap.org/data/2.5/weather', 
                        {'q':[city, state, country],'appid':APIKEY1})
    if response.status_code==200:
        return response.json()['weather'][0]['main']
    else:
        print('! Weather forecast not found')
        pass

cursor.execute('''select customername, contactfirstname, contactlastname, 
                        phone, city, state, country 
                        from customers''')

for row in cursor:
    if get_weather(row[4],row[5],row[6])=='Rain':
        print(f"- It rains: {row[0]}, {row[1]} {row[2]}, {row[3]}")
    else:
        print("- ",row[0])
```
