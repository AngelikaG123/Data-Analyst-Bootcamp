```python
# Create your own method for multiplication 
def multiply(number1, number2):
    return number1 * number2

print('2 * 3 =', multiply(2, 3), '(expected result: 6)')
print('2.5 * 4 =', multiply(2.5, 4), '(expected result: 10.0)')
print('4 * 0.25 =', multiply(4, 0.25), '(expected result: 1.0)')
print('0.4 * 2.5 =', multiply(0.4, 2.5), '(expected result: 1.0)')
print('4 * 3 + 5 =', multiply(4, 3) + 5, '(expected result: 17)')
print('10 + 2 * 6 =', 10 + multiply(2, 6), '(expected result: 22)')
```

```python
# Create your own method for "power" and "root"
def power(base, exponent):
    return base ** exponent

def root(base, exponent = 2):
    return base ** (1/exponent)

print('2 ** 3 =', power(2, 3), '(expected result: 8)')
print('1.5 ** 2 =', power(1.5, 2), '(expected result: 2.25)')
print('2 ** 5 - 30 =', power(2, 5) - 30, '(expected result: 2)')
print('the 2nd degree root of 64 is', root(64, 2), '(expected result: 8.0)')
print('the 2nd degree root of 64 is', root(64), '(expected result: 8.0)')
print('the square root of 2401, minus 5, is', root(2401, 4) - 5, '(expected result: 2)')
```

```python
# Write a function is_even(number) that takes a number and returns True if the number is even, False otherwise.
# Write a function is_odd(number) that takes a number and returns True if the number is odd, False otherwise. 
# You can calculate the result (analogous to the is_even function), or use the is_even function in the is_odd function code.


def is_even(number):
    return (number % 2 == 0)

def is_odd(number):
    return not is_even(number)

if is_even(4):
    print('4 is even')
else:
    print('4 is... odd? Are you sure?')

if is_even(7):
    print('7 is... even? Are you sure?')
else:
    print('7 is odd')

if is_odd(9):
    print('9 is odd')
else:
    print('9 is... even? Are you sure?')

if is_odd(12):
    print('12 is... odd? Are you sure?')
else:
    print('12 is even')

print(is_even(4))
```

```python
# Write a find_short_words function that will accept a list of words. 
# The function should return a list of words shorter than 5 characters.

def find_short_words(lst):
    lista = []
    for x in lst:
        if len(x) < 5:
            lista.append(x)
    return lista
            
test_list = ['Jackdaws', 'love', 'my', 'big', 'sphinx', 'of', 'quartz']
result = find_short_words(test_list)

if find_short_words(test_list) == ['love', 'my', 'big', 'of']:
    print('Well done!')
else:
    print('For the text "Jackdaws love my big sphinx of quartz" the result was expected:')
    print('["love", "my", "big", "of"]')
    print('But the function returned:', result)
```

```python
# Operations on dictionary type

def user_data(first_name, last_name):
    user_dictionary = {
        'first_name' : first_name,
        'last_name' : last_name,
        'full_name' : first_name + " " + last_name,
        'initials' : first_name[0] + ". " + last_name[0] + "."
    }
    return user_dictionary

# Test data
if user_data('Jan', 'Kowalski') == {
        "first_name": "Jan",
        "last_name": "Kowalski",
        "full_name": "Jan Kowalski",
        "initials": "J. K."
}:
    print('Zaliczone!')
else:
    print('The user_data function did not return the expected content. Expected:')
    print('{"first_name": "Jan", "last_name": "Kowalski", "full_name": "Jan Kowalski", "initials": "J. K."}')
    print('The function returned:')
    print(user_data('Jan', 'Kowalski'))
```

```python
# Write a function create_email(first_name, last_name) that will generate an email address in the mysupercompany.com domain.
# The username should be created based on the arguments: first letter of the name, dot, surname - all written in lowercase letters.

def create_email(first_name, last_name):
    email = "{}.{}@mojasuperfirma.com".format(first_name[0].lower(),last_name.lower())
    return email

# Test data
if create_email('Jan', 'Kowalski') == 'j.kowalski@mojasuperfirma.com':
    print('Well done!')
else:
    print('For the "Jan" and "Kowalski" arguments, the function was expected to return j.kowalski@mojasuperfirma.com')
    print('The function returned:', create_email('Jan', 'Kowalski'))

if create_email('Anna', 'Nowak') == 'a.nowak@mojasuperfirma.com':
    print('Well done!')
else:
    print('For the "Anna" and "Nowak" arguments, the function was expected to return a.nowak@mojasuperfirma.com')
    print('The function returned:', create_email('Anna', 'Nowak'))
```

```python
# Write an anonymize(email) function that will take an email address as an argument and return a text consisting of:
# - first three characters of the email address
# - three stars: '***'
# - last five characters of the email address
# If the address is less than 10 characters long, the function should return a text consisting of:
# - three stars: '***'
# - last five characters of the email address


def anonymize(email):
    anonymized_mail = f"{email[:3] if len(email) >= 10 else ''}***{email[-5:]}"
    return anonymized_mail

# Test data
if anonymize('jan.kowalski@microsoft.com') == 'jan***t.com':
    print('Well done!')
else:
    print('For the argument "jan.kowalski@microsoft.com", the function was expected to return jan***t.com')
    print('The function returned:', anonymize('jan.kowalski@microsoft.com'))

if anonymize('jk@xd.pl') == '***xd.pl':
    print('Well done!')
else:
    print('For the argument "jk@xd.pl" the function was expected to return ***xd.pl')
    print('The function returned:', anonymize('jk@xd.pl'))
```

```python
# Create a string reversing function

def reverse_str(text):
    return text[::-1]

# Test data
if reverse_str('ABCDefg') == 'gfeDCBA':
    print('Well done!')
else:
    print('For the "ABCDefg" argument, the function was expected to return gfeDCBA')
    print('The function returned:', reverse_str('ABCDefg'))
```
