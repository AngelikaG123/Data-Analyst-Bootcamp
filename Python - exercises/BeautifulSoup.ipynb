Using https://www.autocentrum.pl/dane-techniczne/, retrieve car brands. Based on that find all models of 10 brands and few details of each one of them.

```python
import requests
from bs4 import BeautifulSoup

def get_brand_pages_urls():
    r = requests.get("https://www.autocentrum.pl/dane-techniczne/")
    soup = BeautifulSoup(r.text, 'html.parser')
    result = soup.find_all('div', class_="make-wrapper")
    link_list = []

    for r in result:
        link = r.find('a', class_="make")
        link2 = link['href']
        ready_link = "https://www.autocentrum.pl"+ link2
        link_list.append(ready_link)
    return link_list

def get_model_pages_urls(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    result = soup.find_all('a', class_="car-selector-box")
    link_list = []

    for r in result:
        link2 = r['href']
        ready_link = "https://www.autocentrum.pl"+ link2
        link_list.append(ready_link)
    return link_list

def get_model_details(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    name_div = soup.find('div', class_='name')
    dictionary = {}
    dictionary["Model"] = name_div.get_text(strip=True)
    test = soup.find_all('div', class_='group')
    test2 = test[0].find_all('span', class_='dt-param-value')
    dictionary["Liczba drzwi"] = test2[0].text
    dictionary["Liczba miejsc"] = test2[1].text

    return dictionary

brand_list = get_brand_pages_urls()
del brand_list[4] 
del brand_list[4] 
del brand_list[4] 
brand_list = brand_list[:10]

for i in brand_list:
    models = get_model_pages_urls(i)
    for i in models:
        try:
            result = get_model_details(i)
            print(f'Model: {result["Model"].ljust(55)} | Number of doors: {result["Liczba drzwi"].ljust(3)} | Number of seats: {result["Liczba miejsc"]}')
        except:
            continue
```
